```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("PS4_GamesSales.csv", encoding='latin1')
```

```python
df['Year'] = df['Year'].fillna(df['Year'].median())
df
```

```python
publisher_mode = df['Publisher'].mode()[0]
```

```python
publisher_mode
```

```python
df['Publisher'] = df['Publisher'].fillna(publisher_mode)
df
```

```python
print(df.isnull().sum())
```

#### Objective 1: Identify the top 5 best-selling games globally

```python
top_5_games = df.nlargest(5, 'Global')[['Game', 'Global']]
print(top_5_games)
```

#### Objective 2: Compare regional sales for each genre

```python
regional_sales = df.groupby('Genre')[['North America', 'Europe', 'Japan', 'Rest of World']].sum()
print(regional_sales)
```

#### Objective 3. Find the Publisher with the Highest Total Global Sales

```python
top_publisher = df.groupby('Publisher')['Global'].sum().idxmax()
top_publisher_sales = df.groupby('Publisher')['Global'].sum().max()
print(f"Top Publisher: {top_publisher}, Total Global Sales: {top_publisher_sales}")
```

#### Objective 4. Analyze the Year with the Highest Number of Game Releases

```python
most_games_year = df['Year'].value_counts().idxmax()
game_count = df['Year'].value_counts().max()
print(f"Year with the Most Releases: {most_games_year}, Total Games: {game_count}")
```

#### Objective 5. Determine the Most Popular Genre Globally Based on Total Sales

```python
most_popular_genre = df.groupby('Genre')['Global'].sum().idxmax()
genre_sales = df.groupby('Genre')['Global'].sum().max()
print(f"Most Popular Genre: {most_popular_genre}, Total Global Sales: {genre_sales}")
```

#### Objective 6. Identify the Game with the Highest Sales in Japan Compared to Other Regions

```python
highest_japan_game = df.nlargest(1, 'Japan')
print(highest_japan_game[['Game', 'Japan', 'North America', 'Europe', 'Rest of World']])
```

#### Objective 7. Examine Correlation Between Sales in North America and Europe

```python
correlation = df[['North America', 'Europe']].corr().iloc[0, 1]
print(f"Correlation between North America and Europe sales: {correlation}")
```

#### Objective 8. Compare Sales Growth Trends for Action and Shooter Genres

```python
action_shooter_trends = df[df['Genre'].isin(['Action', 'Shooter'])].groupby(['Year', 'Genre'])['Global'].sum().unstack()
print(action_shooter_trends)
```

#### Objective 9. Identify the Top Publisher in Terms of Sales for Each Region

```python
top_publishers_region = df.groupby('Publisher')[['North America', 'Europe', 'Japan', 'Rest of World']].sum().idxmax()
print(top_publishers_region)
```

## Objective 10. Visualize Key Data Insights

#### a. Distribution of Global Sales (Histogram)

```python
df['Global'].hist(bins=30, color='skyblue', edgecolor='black')
plt.title('Distribution of Global Sales')
plt.xlabel('Global Sales')
plt.ylabel('Frequency')
plt.show()
```

#### b. Total Sales Comparison for Regions (Bar Chart)

```python
region_sales = df[['North America', 'Europe', 'Japan', 'Rest of World']].sum()
region_sales.plot(kind='bar', color=['blue', 'orange', 'green', 'red'])
plt.title('Total Sales Comparison by Region')
plt.xlabel('Region')
plt.ylabel('Total Sales')
plt.show()
```

#### c. Share of Total Global Sales by Genre (Pie Chart)

```python
genre_sales = df.groupby('Genre')['Global'].sum()
genre_sales.plot(kind='pie', autopct='%1.1f%%', figsize=(8, 8))
plt.title('Share of Total Global Sales by Genre')
plt.ylabel('')  # Hide the y-axis label
plt.show()
```

#### d. Trend of Global Sales Over the Years (Line Chart)

```python
global_sales_trend = df.groupby('Year')['Global'].sum()
global_sales_trend.plot(kind='line', marker='o', color='purple')
plt.title('Trend of Global Sales Over the Years')
plt.xlabel('Year')
plt.ylabel('Global Sales')
plt.show()
```

#### e. Top 10 Best-Selling Games Globally (Horizontal Bar Chart)

```python
top_10_games = df.nlargest(10, 'Global')[['Game', 'Global']]
top_10_games.plot(x='Game', y='Global', kind='barh', color='teal')
plt.title('Top 10 Best-Selling Games Globally')
plt.xlabel('Global Sales')
plt.ylabel('Game')
plt.show()
```

#### f. Correlation Between Regional Sales (Heatmap)

```python
import seaborn as sns
region_corr = df[['North America', 'Europe', 'Japan', 'Rest of World']].corr()
sns.heatmap(region_corr, annot=True, cmap='coolwarm')
plt.title('Correlation Between Regional Sales')
plt.show()
```

#### g. Contribution of Each Genre to Regional Sales (Stacked Bar Chart)

```python
genre_region_sales = df.groupby('Genre')[['North America', 'Europe', 'Japan', 'Rest of World']].sum()
genre_region_sales.plot(kind='bar', stacked=True, figsize=(10, 6), colormap='viridis')
plt.title('Contribution of Each Genre to Regional Sales')
plt.xlabel('Genre')
plt.ylabel('Sales')
plt.show()
```
